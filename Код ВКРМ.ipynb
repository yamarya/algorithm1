```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
def get_signal_length(signal):
      return len(signal)
```

```python
data = pd.read_excel('NIR DATA.xlsx')
start_time = 0
end_time = 25
filtered_data = data[(data['TIME'] >= start_time) & (data['TIME'] <= end_time)]
time = filtered_data['TIME']
r1 = filtered_data['RHEO1']
r2 = filtered_data['RHEO2']
```

# Графики

```python
plt.figure(figsize=(17, 3))
plt.plot(time,r1,color='red',label='RAW')
plt.title("Channel 1")
plt.xlabel("time, sec")
plt.ylabel("R, Ω")
plt.xlim(2,22)
plt.ylim(25.5,28)
```

```python
plt.figure(figsize=(17, 3))
plt.plot(time,r2,color='red',label='RAW')
plt.title("Channel 2")
plt.xlabel("time, sec")
plt.ylabel("R, Ω")
plt.xlim(2,22)
plt.ylim(24,27)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(r1, r2)
plt.xlabel('first channel')
plt.ylabel('second channel')
plt.title('phase portrait')
#plt.xlim(-0.3,0.3)
#plt.ylim(-0.3,0.3)
plt.grid(True)
plt.show()
```

# Экспоненциальное усреднение

```python
a1 = 0.03125
r11 = [r1[0]*a1]
for i in range(1,get_signal_length(filtered_data)):
   r11.append(a1*r1[i]+(1-a1)*r11[i-1])
```

```python
plt.figure(figsize=(17, 3))
#plt.plot(time,r1,color='red',label='RAW')
plt.plot(time,r11,color='blue',label='Filtered')
plt.title("Channel 1")
plt.xlabel("time, sec")
plt.ylabel("R, Ω")
plt.legend()
plt.xlim(2,22)
plt.ylim(25.5,28)
```

```python
a2 = 0.03125
r22 = [r2[0]*a2]
for i in range(1,get_signal_length(filtered_data)):
    r22.append(a2*r2[i]+(1-a2)*r22[i-1])
```

```python
plt.figure(figsize=(17, 3))
#plt.plot(time,r2,color='red',label='RAW')
plt.plot(time,r22,color='blue',label='Filtered')
plt.title("Channel 2")
plt.xlabel("time, sec")
plt.ylabel("R, Ω")
plt.legend()
plt.xlim(2,22)
plt.ylim(24,27)
```

# Удаление постоянной составляющей

```python
r111 = [r11[0]]
for i in range(1,get_signal_length(filtered_data)):
    r111.append(r11[i]+0.95*r111[i-1])
```

```python
r1111 = [r111[0]]
for i in range(1,get_signal_length(filtered_data)):
    r1111.append(r111[i] - r111[i-1])
```

```python
plt.figure(figsize=(17, 3))
plt.plot(time,r1111)
plt.title("Channel 1")
plt.xlabel("time, sec")
plt.ylabel("R, Ω")
plt.xlim(2,22)
plt.ylim(-0.5,0.5)
```

```python
r222 = [r22[0]]
for i in range(1,get_signal_length(filtered_data)):
    r222.append(r22[i]+0.95*r222[i-1])
```

```python
r2222 = [r222[0]]
for i in range(1,get_signal_length(filtered_data)):
    r2222.append(r222[i] - r222[i-1])
```

```python
plt.figure(figsize=(17, 3))
plt.plot(time,r2222)
plt.title("Channel 2")
plt.xlabel("time, sec")
plt.ylabel("R, Ω")
plt.xlim(2,22)
plt.ylim(-0.5,0.5)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(r1111, r2222)
plt.xlabel('first channel')
plt.ylabel('second channel')
plt.title('phase portrait')
plt.xlim(-0.3,0.3)
plt.ylim(-0.3,0.3)
plt.grid(True)
plt.show()
```
